```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def floyd_steinberg_dithering(image):
    img = image.convert("L")
    pixels = np.array(img, dtype=float)
    rows, cols = pixels.shape

    for y in range(rows):
        for x in range(cols):
            old_pixel = pixels[y, x]
            new_pixel = 255 if old_pixel > 127 else 0
            pixels[y, x] = new_pixel
            quant_error = old_pixel - new_pixel
            if x + 1 < cols:
                pixels[y, x + 1] += quant_error * 7 / 16.0
            if y + 1 < rows:
                if x - 1 >= 0:
                    pixels[y + 1, x - 1] += quant_error * 3 / 16.0
                pixels[y + 1, x] += quant_error * 5 / 16.0
                if x + 1 < cols:
                    pixels[y + 1, x + 1] += quant_error * 1 / 16.0

    pixels = np.clip(pixels, 0, 255)
    return Image.fromarray(pixels.astype(np.uint8))

def jarvis_judice_ninke_dithering(image):
    img = image.convert("L")
    pixels = np.array(img, dtype=float)
    rows, cols = pixels.shape


    kernel = np.array([[0, 0, 0, 7, 5],
                       [3, 5, 7, 5, 3],
                       [1, 3, 5, 3, 1]]) / 48.0

    for y in range(rows):
        for x in range(cols):
            old_pixel = pixels[y, x]
            new_pixel = 255 * (old_pixel // 128)
            pixels[y, x] = new_pixel
            quant_error = old_pixel - new_pixel
            for ky in range(3):
                for kx in range(5):
                    ny = y + ky
                    nx = x + kx - 2
                    if ny < rows and 0 <= nx < cols:
                        pixels[ny, nx] += quant_error * kernel[ky, kx]

    return Image.fromarray(np.clip(pixels, 0, 255).astype(np.uint8))


#I am loading an image of tiger from the folder
image_path = 'tiger.jpeg'
image = Image.open(image_path)

# Calling floyd_steinberg_dithering to apply on my image
fs_dithered_image = floyd_steinberg_dithering(image)

#  Applying Jarvis-Judice-Ninke dithering to the image
jjn_dithered_image = jarvis_judice_ninke_dithering(image)

# Comparing the image
plt.figure(figsize=(10, 5))

# Displaying Floyd-Steinberg Dithered Image for the original image
plt.subplot(1, 3, 2)
plt.imshow(fs_dithered_image.convert("RGB"))
plt.title("Floyd-Steinberg Dithering Filter")
plt.axis('off')

# Displaying Jarvis-Judice-Ninke Dithered Image for the original image
plt.subplot(1, 3, 3)
plt.imshow(jjn_dithered_image.convert("RGB"))
plt.title("Jarvis-Judice-Ninke Dithering Filter")
plt.axis('off')

plt.tight_layout()
plt.show()


```
